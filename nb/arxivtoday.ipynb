```python
# ライブラリのインストール
# !pip install arxiv pandas

import arxiv
import pandas as pd
from datetime import datetime, timedelta

# 検索キーワードを設定
# ユーザーの興味関心領域である「意識」「自己」「デジタルゴースト」などに関連するキーワード
search_keywords = [
    "consciousness",
    "self-awareness",
    "digital ghost",
    "embodied intelligence",
    "brain-computer interface",
    "machine learning"
]
query = " OR ".join([f'"{keyword}"' for keyword in search_keywords])

# 検索クライアントの作成
client = arxiv.Client()

# 論文を検索
# 過去1日以内(timedelta(days=1))に更新された論文を対象
search = arxiv.Search(
  query = query,
  max_results = 10,
  sort_by = arxiv.SortCriterion.LastUpdatedDate
)

results = client.results(search)

# 結果をリストに格納
papers_list = []
for result in results:
    # 昨日の日付以降に更新された論文のみを抽出
    yesterday = datetime.now().date() - timedelta(days=10)
    if result.updated.date() >= yesterday:
        papers_list.append({
            "Title": result.title,
            "Authors": ", ".join([author.name for author in result.authors]),
            "Published": result.published.strftime('%Y-%m-%d'),
            "Updated": result.updated.strftime('%Y-%m-%d'),
            "Abstract": result.summary.replace("\n", " "),
            "URL": result.pdf_url
        })

# pandas DataFrameで結果を表示
if papers_list:
    df = pd.DataFrame(papers_list)
    print(f"新着論文が {len(df)} 件見つかりました。")
    display(df)
else:
    print("指定されたキーワードに一致する新着論文はありませんでした。")
```
